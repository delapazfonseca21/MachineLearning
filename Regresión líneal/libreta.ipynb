```python
#Caracteristicas (X), etiquetas (y)
x = datos["Celsius"]
y = datos["fahrenheit"]
```

```python
x
```

```python
y
```

```python
#Realizar transformación para que los datos sean manejados por modelos
#Transformar de arreglo [-40, -10, 0, 8 ...]
#A matriz (arreglo con múltiples arreglos)
#Ej: [[-40],[-10],[0],[8]]
x.values
```

```python
X_procesada = x.values.reshape(-1,1)
Y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
#Creación del modelo
modelo = LinearRegression()
```

```python
#Entrenamiento del modelo (alimentarlo con datos)
modelo.fit(X_procesada, Y_procesada)
```

```python
modelo.predict([[34]])
#De acuerdo a los datos que le di al modelo, predicirá su valor en fahrenheit
```

```python
modelo.score(X_procesada, Y_procesada)
```
