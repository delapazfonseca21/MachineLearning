```python
import pandas as pd
datos = pd.read_csv("datos_practica.csv")
```

```python
import seaborn as sb
sb.scatterplot(x="altura", y="peso", data=datos,
               hue="altura", palette="coolwarm")
```

```python
#Seccionamiento de datos en Carácteristicas(X) y etiquetas (y)
X = datos["altura"]
y = datos["peso"]
```

```python
X
```

```python
y
```

```python
X.values
#El modelo requiere una matriz, es decir, un arreglo con arreglos de una posición
```

```python
X_matriz = X.values.reshape(-1,1)
y_matriz = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
#Creación del modelo
modelo = LinearRegression()
```

```python
#Entrenando al modelo
modelo.fit(X_matriz, y_matriz)
```

```python
modelo.predict([[178]])
```

```python
modelo.score(X_matriz, y_matriz)
```
